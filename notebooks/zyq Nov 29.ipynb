## Training set from MIRA Lab 

```python
# Training data

train_df = pd.read_json('../data/train_data.json')
train_df.replace({"": np.nan}, inplace=True)
display(train_df.head())
train_df.info()
```

```python
# Validation data
# Note: Missing values of 'fit' need to be predicted and filled in. Here we just drop user feedback to meet the format of the testing data.

val_df = train_df.drop(
    columns=['review_summary', 'review', 'rating', 'user_name'])
val_df['fit'].replace({
    'Small': 1,
    'True to Size': 2,
    'Large': 3
},
                      inplace=True)
display(val_df.head())
val_df.info()
```
